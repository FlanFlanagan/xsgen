# Cross Section Generator Example

This notebook demonstrates an example LWR library, called 'brightlite0'. This library was generated using [xsgen command line utility](http://bright-dev.github.io/xsgen/).  XSGen uses [OpenMC](https://mit-crpg.github.io/openmc/) and ORIGEN to calculate reactor specific parameters such as burnup, criticality, transmutation matrices, and cross sections.  These libraries are usable in the Bright-lite reactor facility archetype inside of [Cyclus](http://fuelcycle.org/) simulations.

This notebook assumes that the generated library is installed into the Bright-lite path, nominally '~/.local/share/brightlite'.

First, let's load in some needed tools.

```python
import os

import matplotlib.pyplot as plt
%matplotlib inline
plt.style.use("ggplot")
import pandas as pd
import numpy as np

libname = 'brightlite0'
output_dir = os.path.expanduser("~/.local/share/brightlite/" + libname)
fuel_txt = os.path.join(output_dir, "fuel.txt")
```

### Load in fuel.txt

Next, let's load in fuel data.

```python
fuel = pd.read_csv(fuel_txt, delim_whitespace=True, index_col=0).transpose()
```

### U-235 & U-238 Concentrations

Now let's look at a plot of the concentrations U-235 and U-238 as a function of the residence time in the core.

```python
plt.figure()
fuel.U235.plot(logy=True, figsize=(18, 9))
fuel.U238.plot(logy=True, figsize=(18, 9))
plt.xlabel('Time [days]')
plt.ylabel('Mass [g]')
plt.legend(loc=0)
```

### Burnup and Criticality

Now let's view in the burnup and criticality curves.

```python
plt.figure()
k = (fuel.NEUT_PROD / fuel.NEUT_DEST)
k.name = 'k'
k.plot(figsize=(18, 9))
plt.ylabel('Multiplication Factor')
plt.legend(loc=2)
fuel.BUd.cumsum().plot(figsize=(18, 9), secondary_y=True)
plt.ylabel('Burnup [MWd/kg]')
plt.legend(loc=1)
plt.xlabel('Time [days]')
```

## Cyclus

Now that we have inspected the library and everything looks reasonable, let's run this inside of a Cyclus simulation. A valid input file is provided under the name 'cyclus-example.json'.

First, we'll change the name of the library to what is listed above.

```python
with open('cyclus-example.json', 'r') as f:
    inp = f.read()
inp = inp.replace('brightlite0', libname)
with open('example.json', 'w') as f:
    f.write(inp)
```

Now let's run the file!

```python
if os.path.isfile('example.h5'):
    os.remove('example.h5')
!cyclus -o example.h5 example.json
```
