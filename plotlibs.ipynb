```python
import matplotlib.pyplot as plt
import pandas as pd
import os.path

%matplotlib inline

output_dir = os.path.expanduser("~/xslibs/build-lwr1g/")
outfile_path = os.path.join(output_dir, "brightlite1-fuel.out")
```

```python
data = pd.read_csv(outfile_path, delim_whitespace=True, index_col=0).transpose()
plt.figure()
plt.style.use("ggplot")
data.drop(["NEUT_PROD", "NEUT_DEST", "BUd"], 1).plot(
    loglog=True, legend=True, figsize=(18, 12))
```
